```python
import requests

path="/api/events"
base_url = "http://0.0.0.0:8080"

endpoint = f"{base_url}{path}"
res = requests.get(endpoint)
print(res)
```

```python
import json

path="/api/events"
base_url = "http://0.0.0.0:8080"

endpoint = f"{base_url}{path}"

data=json.dumps({"page":"1"})
res = requests.post(endpoint, data=data)

if res.ok:
    print(res.json())
```
