```python
import requests
from pprint import pprint
```

```python
path = "/api/events/"
base_url = "http://0.0.0.0:8080" # 127.0.0.1
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"page": "asda", "description": '22'}) 
print(response.headers)
print('ok', response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
```

```python
path = "/api/events/"
base_url = "http://0.0.0.0:8080" # 127.0.0.1
create_endpoint = f"{base_url}{path}"

response = requests.get(create_endpoint) 
print(response.headers)
print('ok', response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
```
